```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a dataframe
file_path = 'Airbnb_Open_Data (2).csv'
df = pd.read_csv(file_path)

# Clean the dataset by removing any unnecessary characters from the 'price' column and converting it to a numerical format.
def clean_price_column(df):
    # Remove dollar signs and commas, then convert to float
    df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
    return df

# Apply the cleaning function to the dataframe
df = clean_price_column(df)

# Set the aesthetic style of the plots
sns.set_style("whitegrid")
# 2. Distribution of prices
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='price', bins=50, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price ($)')
plt.ylabel('Count')
plt.xlim(0, 1000)  # Limit x-axis to focus on the main distribution
plt.tight_layout()
plt.show()


# Create a boxplot to visualize the distribution of prices across different neighborhoods
#plt.figure(figsize=(12, 6))
#sns.boxplot(x='neighbourhood', y='price', data=df)
#plt.title('Price Distribution by Neighborhood')
#plt.xticks(rotation=45)
#plt.show()

# Create a boxplot to visualize the distribution of prices across different room types
#plt.figure(figsize=(10, 6))
#sns.boxplot(x='room type', y='price', data=df)
#plt.title('Price Distribution by Room Type')
#plt.xticks(rotation=45)
#plt.show()

# Create a scatter plot to visualize the relationship between construction year and price
#plt.figure(figsize=(10, 6))
#sns.scatterplot(x='Construction year', y='price', data=df)
#plt.title('Price vs. Construction Year')
#plt.xticks(rotation=45)
#plt.show()

# Create a violin plot to visualize the distribution of prices across different neighborhoods
#plt.figure(figsize=(12, 6))
#sns.violinplot(x='neighbourhood', y='price', data=df)
#plt.title('Violin Plot of Price Distribution by Neighborhood')
#plt.xticks(rotation=45)
#plt.show()
# Create a heatmap to visualize the correlation between numerical features
plt.figure(figsize=(10, 8))
correlation_matrix = df[['price', 'Construction year', 'number of reviews', 'availability 365']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
