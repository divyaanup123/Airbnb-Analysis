```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a DataFrame with approximate data from the graph
years = np.arange(2002.5, 2023, 0.5)
prices = [600, 550, 500, 700, 800, 750, 600, 550, 500, 450, 400, 350, 300, 250, 200, 
          300, 400, 500, 600, 700, 800, 900, 950, 900, 850, 800, 750, 700, 650, 600, 
          550, 500, 450, 400, 350, 300, 250, 200, 250, 300, 350]

df = pd.DataFrame({'Year': years, 'Price': prices})

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Price'], marker='o')
plt.axhline(y=600, color='r', linestyle='--', label='Average Price')
plt.title('Price vs Construction Year (Detailed Analysis)')
plt.xlabel('Construction Year')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()

# Calculate some statistics
print(f"Minimum price: ${df['Price'].min()}")
print(f"Maximum price: ${df['Price'].max()}")
print(f"Average price: ${df['Price'].mean():.2f}")
print(f"Price standard deviation: ${df['Price'].std():.2f}")

# Identify years with significant changes
df['Price_Change'] = df['Price'].diff()
significant_changes = df[abs(df['Price_Change']) > 100]
print("\nYears with significant price changes:")
print(significant_changes[['Year', 'Price', 'Price_Change']])
```
